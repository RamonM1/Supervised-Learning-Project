Initial thoughts
    I noticed that there are a mixture of string, numerical, and object data. We need to get counts of the data and look at the distribution of values tha we have, seeing how these features correlate/compare to our target variable.

Univariate Analysis

```python
# Assuming 'amount' as a numerical feature
plt.figure(figsize=(10, 6))
sns.histplot(random_sample['amount'], bins=50, kde=False)
plt.title('Distribution of Transaction Amounts')
plt.xlabel('Amount')
plt.yscale("log")
plt.ylabel('Frequency')
plt.show()
```

```python
#We are going to create a list of the features we are going to use in the analysis and then we will use a for loop to create a histogram for each feature
features = ['step', 'type', 'amount','isFraud','newbalanceOrig','newbalanceDest' ]

for i in features:
	plt.figure(figsize=[8, 6])
	sns.histplot(random_sample[i], bins= 60, kde = False, color = "green")
	plt.xlabel(i, fontsize = 14)
	plt.ylabel("Count")
	plt.yscale("log")
	plt.show()
```

Bivariate Analysis

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='amount', y='oldbalanceOrg',hue='isFraud', data=random_sample, alpha=0.6)
plt.title('Amount vs Old Balance Org')
plt.xlabel('Amount')
plt.ylabel('Old Balance Org')
plt.show()
```

```python
# We are going to create a scatter plot to see the relationship between the amount and new balance origin
sns.scatterplot(x='newbalanceOrig', y='amount', data=random_sample, hue='isFraud')
plt.title('Amount vs New Balance Origin')
plt.yscale('log')
```

```python
# We are going to create a boxplot to see the relationship between the amount and step
sns.boxenplot(x='isFraud', y='step', data=random_sample, hue='type')
plt.legend(loc='upper left')
plt.xlabel('Fraud or Not')
plt.ylabel('Step')
plt.title('Fraud vs Step')
plt.show()
```
